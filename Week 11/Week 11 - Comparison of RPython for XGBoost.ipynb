```python
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier

df = pd.read_csv("dfdata.csv")
df.head()
```

```python
X = df.drop("outcome", axis=1)
y = df["outcome"]

# Function to evaluate model
def evaluate_xgboost(X, y, size):
    if len(X) < size:
        # Duplicate data to match desired size
        repeats = int(np.ceil(size / len(X)))
        X = pd.concat([X]*repeats, ignore_index=True)[:size]
        y = pd.concat([y]*repeats, ignore_index=True)[:size]
    else:
        X = X.iloc[:size]
        y = y.iloc[:size]

    model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    start = time.time()
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    end = time.time()

    print(f"Method: XGBoost via scikit-learn")
    print(f"Dataset size: {size}")
    print(f"Mean accuracy (5-fold CV): {scores.mean():.4f}")
    print(f"Time taken: {end - start:.2f} seconds")
    print("-" * 40)

# Test with various dataset sizes
for size in [100, 1000, 10000, 100000, 1000000]:
    evaluate_xgboost(X, y, size)
```
