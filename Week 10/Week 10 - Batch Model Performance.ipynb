```python
import matplotlib.pyplot as plt

models = ['Dummy', 'KNN-1', 'KNN-2', 'KNN-3-1', 'KNN-3-2']
steps = ['Dummy', 'KNN Training', 'KNN Prediction', 'KNN Retraining', 'KNN Retrained Prediction']
train = [0.4767, 0.8983, 0.9717, 0.8983, 0.9717]
accuracy = [0.5075, 0.895, 0.963, 0.721, 0.965]
recall = [0.5048, 0.8867, 0.9407, 0.4163, 0.913]
precision = [0.5327, 0.9045, 0.9061, 1.0, 0.8873]
```

```python
x = list(range(len(models)))
plt.figure(figsize=(10, 6))

plt.plot(x, train, 'o-', label='Train', linewidth=2, markersize=8)
plt.plot(x, accuracy, 's-', label='Accuracy', linewidth=2, markersize=8)
plt.plot(x, recall, '^-', label='Recall', linewidth=2, markersize=8)
plt.plot(x, precision, 'd-', label='Precision', linewidth=2, markersize=8)

plt.xlabel('Model Steps')
plt.ylabel('Score')
plt.title('Batch Model Performance Across Different Steps')
plt.xticks(x, models)
plt.grid(True, linestyle='--', alpha=0.7)
ax1 = plt.gca()
ax2 = ax1.twiny()
ax2.set_xticks(x)
ax2.set_xticklabels(steps)
ax2.set_xlabel('Process Steps')
plt.ylim(0.4, 1.05)
plt.legend(loc='lower right')
plt.annotate('Major drop in recall', xy=(3, 0.4163), xytext=(3.2, 0.3),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))
plt.annotate('Perfect precision', xy=(3, 1.0), xytext=(3.2, 0.95),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1.5))

plt.tight_layout()
plt.show()
```
